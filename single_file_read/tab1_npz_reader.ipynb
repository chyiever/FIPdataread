```python
data1 = payload["phase_data"]
# 带通滤波
import scipy.signal as signal
import numpy as np

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.butter(order, [low, high], btype="band")
    return signal.filtfilt(b, a, data)
fs = 200000  # 采样率
lowcut = 5000  # 低频截止
highcut = 50000  # 高频截止
filtered_data1 = bandpass_filter(data1, lowcut, highcut, fs)[2000000:6000000]
t =  np.arange(len(filtered_data1   )) / fs # 时间向量


# Plot the original time-domain waveform
import matplotlib.pyplot as plt
# 字体设置
plt.rcParams["font.family"] = "Times New Roman"
%matplotlib qt
plt.figure(figsize=(12, 5))
plt.plot(t, filtered_data1, color="red", linewidth=0.5)
plt.grid()
plt.xlabel("Time (s)", fontsize=16)
plt.ylabel("Amplitude", fontsize=16)
plt.tick_params(labelsize=14)
plt.title("Filtered Time-Domain Waveform", fontsize=14)
```
